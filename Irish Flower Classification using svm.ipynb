```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
import pandas as pd
```

Import iris flowers dataset

```python
def getData():
    iris = datasets.load_iris()
    return iris
iris = getData()
```

iris.data [ : , : 2 ]

The first slice before the comma (:): This refers to rows. Using : alone means selecting all rows in the array.

The second slice after the comma (:2): This refers to columns. :2 means selecting columns from index 0 up to (but not including) index 2.

The iris.target array contains integers that correspond to the different species of iris flowers. Each integer label corresponds to a specific species as follows:
0: Setosa
1: Versicolor
2: Virginica

```python
def get2DPlot(iris):
    X = iris.data[:, :2] 
    Y = iris.target # 0 1 2 
    
    X_min, X_max = X[:, 0].min() - .5, X[:, 0].max() + .5 #Add 0.5 to both positive and negative directions                                                     
    Y_min, Y_max = X[:, 1].min() - .5, X[:, 1].max() + .5 #for a cleaner visualization.
    
    plt.figure(2, figsize=(8, 6)) #create figure with 2 dimensions and 8 inches wide and 6 inches tall
    plt.clf() #clf stand for clear figure allow you create new figure without content of the previous figure

    
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired)
    #cmap for color map
    #c for class
    # X[:,0] array contain Sepal length
    # X[:,1] array contain Sepal width
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')

    plt.xlim(X_min, X_max)
    plt.ylim(Y_min, Y_max)
    plt.xticks(())
    plt.yticks(())
    plt.show()

    
```

```python
get2DPlot(iris)
```

```python
from matplotlib.colors import ListedColormap

# Create a meshgrid of points
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Define the SVM models to visualize
models = [svc, rbf_svc, poly_svc, lin_svc]
titles = ['SVC with Linear Kernel',
          'SVC with RBF Kernel',
          'SVC with Polynomial (degree 3) Kernel',
          'LinearSVC']

# Create a figure with subplots
plt.figure(figsize=(12, 8))
for i, clf in enumerate(models):
    plt.subplot(2, 2, i + 1)

    # Plot the decision boundary using contour plots
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

    # Plot the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])

plt.tight_layout()
plt.show()

```
